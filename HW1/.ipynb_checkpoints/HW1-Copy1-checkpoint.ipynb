# Question 3

First, we would like to standardize all non-categorical values, so that Euclidian distance may be calculated more effectivly.

```python
from sklearn.preprocessing import StandardScaler
# scale the data for SVC
x_train_standard = x_train.copy()
x_train_standard[['tenure','MonthlyCharges']] = StandardScaler().fit_transform(x_train[['tenure','MonthlyCharges']])
```

Then, we use gridsearch to identify which parameters are best for SVC. The hyperparameters tuned are: kernels, gamma, C and degree.

```python
x_test_standard = x_test.copy()
x_test_standard[['tenure','MonthlyCharges']] = StandardScaler().fit_transform(x_test[['tenure','MonthlyCharges']])
```

```python
from sklearn.svm import SVC
# fit a model with optimal parameters
# calculate the accuracy
score = 0
poly = 0
for i in range(6): 
    clf = SVC(degree = i+1 , kernel = 'poly',gamma = 'auto')
    clf.fit(x_train_standard,y_train)
    y_pred = clf.predict(x_test_standard)
    new_score = accuracy_score(y_test,y_pred)
    if new_score > score:
        score = new_score
        ploy = i + 1
print(score,ploy,sep = ' ')
        
```

polynomial kernel with degree 1 is the linear kernel

```python
# fit a model with rbf kernel
# calculate the accuracy

clf = SVC(kernel = 'rbf',gamma = 'auto')
clf.fit(x_train_standard,y_train)
y_pred = clf.predict(x_test_standard)
accuracy_score(y_test,y_pred)
        
```

The SVC with linear,ploy and rbf kernel returns a best estimation of accuracy around 0.809
