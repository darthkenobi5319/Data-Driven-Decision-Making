```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

```python
# synthetic dataset for simple regression
from sklearn.datasets import make_regression
plt.figure()
plt.title('Sample regression problem with one input variable')
X_R1, y_R1 = make_regression(n_samples = 100, n_features=1,
                            n_informative=1, bias = 150.0,
                            noise = 30, random_state=0)
plt.scatter(X_R1, y_R1, marker= 'o', s=50)
plt.show()
```

```python
# synthetic dataset for more complex regression
from sklearn.datasets import make_friedman1
plt.figure()
plt.title('Complex regression problem with one input variable')
X_F1, y_F1 = make_friedman1(n_samples = 100,
                           n_features = 7, random_state=0)

plt.scatter(X_F1[:, 4], y_F1, marker= 'o', s=50)
plt.show()
```

```python
# train and test a simple linear regression model
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(X_R1, y_R1,
                                                   random_state = 0)
linreg = LinearRegression().fit(X_train, y_train)

print('linear model coeff (w): {}'
     .format(linreg.coef_))
print('linear model intercept (b): {:.3f}'
     .format(linreg.intercept_))
print('R-squared score (training): {:.3f}'
     .format(linreg.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linreg.score(X_test, y_test)))
```

```python
#Linear regression: example plot
plt.figure(figsize=(5,4))
plt.scatter(X_R1, y_R1, marker= 'o', s=50, alpha=0.8)
plt.plot(X_R1, linreg.coef_ * X_R1 + linreg.intercept_, 'r-')
plt.title('Least-squares linear regression')
plt.xlabel('Feature value (x)')
plt.ylabel('Target value (y)')
plt.show()
```

```python
print('Model result')
print('linear model intercept: {}'
     .format(linreg.intercept_))
print('linear model coeff:\n{}'
     .format(linreg.coef_))
print('R-squared score (training): {:.3f}'
     .format(linreg.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linreg.score(X_test, y_test)))
```

```python
# Lasso regression
df = pd.read_csv('data/20scores.csv')
df.head()
```

```python
x_cols = df.columns[1:]
X = df[x_cols]

y = df['revenue']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 10)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso(normalize=True, random_state=10)
lasso_model.fit(X_train,y_train)
```

```python
lasso_model.coef_
```

```python
from sklearn.metrics import mean_squared_error

predictions = model.predict(X_test)
print(mean_squared_error(predictions, y_test)**0.5)
```

```python
lasso_predictions = lasso_model.predict(X_test)
print(mean_squared_error(lasso_predictions, y_test)**0.5)
```
